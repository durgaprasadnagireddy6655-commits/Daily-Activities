**Feature Encoding**

Feature encoding is the process of converting a non numerical data into a numerical format that machine learning algorithms can understand and process.

we choose this because:: the type of categorical data, the number of unique values in a categorical column, and the model we plan to train Types of categorical data--> nominal(no order), ordinal(order)

Nominal Data: No order existence ex: red, ordinal data order exists ex: size(small<medium<large) rating(low<medium<High) Importance: text data can't be used directly in model training. proper encoding: Improves model Accuracy Prevents wrong assumptions Make training stable

**Common Feature Encoding Techniques(7)**

Label encoding, One hot encoding , Target Encoding, Ordinal Encoding , Frequency Encoding, Binary Encoding, Feature Hashmap

**Label encodingt**
 Each category is assigned a unique interger( e.g: "red":0,"green":1)
 use for ordinal data ( inherent order) or with tree based models that are not sensitive to the numberical order.

```python
import pandas as pd
df=pd.DataFrame({
    "color": ["Red", "Green", "Blue","Green","Red"],
    "Size": ["Small", "medium","Large","medium","Small"],
    "City": ["Hyderabad","Delhi","Mumbai","Delhi","Chennai"],
    "Target": [10,20,15,25,12]
})
df
```

```python
oh=pd.get_dummies(df,columns=["color"])
oh
```

**ordinary encoding**

  Similar to Label Encoding, but the integer assignment is based on the specific, predefined order or rank of the categories( e.g: "low":0, "medium":1,"high":2)
  
  Use specifically when the categorical variable has the natural meaning ful order.


```python
from sklearn.preprocessing import OrdinalEncoder # takes the values based on order
```

```python
order=[["Small","medium","Large"]]

df["Size_ordinal"]=OrdinalEncoder(categories=order).fit_transform(df[["Size"]])
df[["Size","Size_ordinal"]]
```

**Target Encoding**
Each category is replaced with the mean of the target varibale for that category.

Effective for high-cardinality features and when there's a strong relationship between the category and the target variable, but requires careful validation to valide target values

```python
te=df.groupby("City")["Target"].mean()
df["city_Target_Enc"]=df["City"].map(te)
df[["City","city_Target_Enc"]]
```

**Frequency Encoding**

Replaces each category with its frequency(count) in the dataset.

   Useful when the frequencey of a category is predictive of the target variable.

```python
fe=df.groupby("City")["Target"].count()
df["city_Target_Enc"]=df["City"].map(fe)  #method 1
df[["City","city_Target_Enc"]]
```

```python
f=df["City"].value_counts()
df["city_Target_Enc"]=df["City"].map(fe) # method 2
df[["City","city_Target_Enc"]]
```
